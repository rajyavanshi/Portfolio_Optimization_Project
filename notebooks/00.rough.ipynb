```python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import os

#  Update your actual folder paths here
base_dir = "D:/Portfolio Optimzation project"
raw_path = os.path.join(base_dir, "raw_data", "3MINDIA.csv")           
processed_path = os.path.join(base_dir, "processed_data", "3MINDIA_daily.csv")

# Load CSVs
raw_df = pd.read_csv(raw_path)
processed_df = pd.read_csv(processed_path)

# columns and number of rows
raw_info = {"File": "Raw Data", "Columns": list(raw_df.columns), "Number of Rows": len(raw_df)}
processed_info = {"File": "Processed Data", "Columns": list(processed_df.columns), "Number of Rows": len(processed_df)}
summary_df = pd.DataFrame([raw_info, processed_info])

print("\n=== Comparison Summary ===")
print(summary_df.to_string(index=False))
display(summary_df)

#  common numeric columns
common_columns = list(set(raw_df.columns) & set(processed_df.columns))
numeric_common_columns = [col for col in common_columns 
                          if pd.api.types.is_numeric_dtype(raw_df[col]) 
                          and pd.api.types.is_numeric_dtype(processed_df[col])]

print("\nCommon numeric columns available for comparison:", numeric_common_columns)

# Plot comparison for first few numeric columns
for col in numeric_common_columns[:5]:  # Limit to 5 for readability
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(raw_df[col], label=f"Raw {col}")
    plt.title(f"Raw Data - {col}")
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(processed_df[col], label=f"Processed {col}", color='orange')
    plt.title(f"Processed Data - {col}")
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```

| -------------------------- | -------------------------------------------------- | ---------------------------------------------- |

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os, random

base_dir = "D:/Portfolio Optimzation project"
raw_path = os.path.join(base_dir, "raw_data", "3MINDIA.csv")           
processed_path = os.path.join(base_dir, "processed_data", "3MINDIA_daily.csv")

raw_df = pd.read_csv(raw_path)
processed_df = pd.read_csv(processed_path)

# Full comparison function
def compare_datasets(raw_df, processed_df, raw_name="Raw Data", processed_name="Processed Data"):
    print(f" Comparing {raw_name} vs {processed_name}\n")
    print(f"{raw_name} shape: {raw_df.shape}")
    print(f"{processed_name} shape: {processed_df.shape}\n")
    
    print(f"{raw_name} columns:\n{list(raw_df.columns)}\n")
    print(f"{processed_name} columns:\n{list(processed_df.columns)}\n")

    # Correlation matrices comparison
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.heatmap(raw_df.corr(numeric_only=True), annot=False)
    plt.title(f"{raw_name} Correlation Matrix")

    plt.subplot(1, 2, 2)
    sns.heatmap(processed_df.corr(numeric_only=True), annot=False)
    plt.title(f"{processed_name} Correlation Matrix")
    plt.tight_layout()
    plt.show()

    # Histograms
    numeric_cols = [col for col in raw_df.columns if pd.api.types.is_numeric_dtype(raw_df[col])]
    for col in numeric_cols[:5]:
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.hist(raw_df[col].dropna(), bins=50, alpha=0.7)
        plt.title(f"{raw_name} - {col} Distribution")

        plt.subplot(1, 2, 2)
        if col in processed_df.columns:
            plt.hist(processed_df[col].dropna(), bins=50, alpha=0.7, color='orange')
            plt.title(f"{processed_name} - {col} Distribution")
        plt.tight_layout()
        plt.show()

    # Time series visualization for price-like columns
    price_cols = [col for col in ["Close", "Adj Close", "Open", "High", "Low"] if col in raw_df.columns]
    for col in price_cols:
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(raw_df[col])
        plt.title(f"{raw_name} - {col} Trend")

        plt.subplot(1, 2, 2)
        if col in processed_df.columns:
            plt.plot(processed_df[col], color='orange')
            plt.title(f"{processed_name} - {col} Trend")
        plt.tight_layout()
        plt.show()

    # Volatility (rolling std) visualization
    if "Close" in raw_df.columns and "Close" in processed_df.columns:
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(raw_df["Close"].rolling(window=10).std(), label="10-day Rolling Vol")
        plt.title(f"{raw_name} - Rolling Volatility")
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(processed_df["Close"].rolling(window=10).std(), label="10-day Rolling Vol", color='orange')
        plt.title(f"{processed_name} - Rolling Volatility")
        plt.legend()
        plt.tight_layout()
        plt.show()

# Execute comparison
compare_datasets(raw_df, processed_df)

# Function for random CSV pair selection
def pick_random_csv_pair(raw_folder, processed_folder):
    raw_files = [f for f in os.listdir(raw_folder) if f.endswith(".csv")]
    chosen_raw = random.choice(raw_files)
    processed_match = chosen_raw.replace(".csv", "_daily.csv")
    processed_path = os.path.join(processed_folder, processed_match)
    raw_path = os.path.join(raw_folder, chosen_raw)

    if os.path.exists(processed_path):
        print(f" Selected Pair:\nRaw: {chosen_raw}\nProcessed: {processed_match}")
        return raw_path, processed_path
    else:
        print(f" No matching processed file found for {chosen_raw}")
        return None, None
```
